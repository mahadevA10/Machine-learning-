# 1) Plot the curves/planes on python for the given expressions mentioned below:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,101)
a = 0*x+2.4
b = 3*x+5
c = -2.5*x + 10
d = x+11
# print(x,a,b,c,d)
```

```python
plt.figure(figsize=(10,10))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.plot(x,a,'black',label='a) f(x) = 2.4')
plt.plot(x,b,'blue',label='b) f(x) = 3x + 5')
plt.plot(x,c,'red',label='c) f(x) = –2.5x + 10')
plt.plot(x,d,'yellow',label='d) f(x) = x+11')
plt.legend()
plt.show()
```

#e) Write a function to check whether the point (–1, 3) is bounded with in the area enclosed by the curves of a), b), c) and d).


---



---



```python
def checkPoint1(x1,y1):
  res = False
  a1 = 0*x1+2.4
  b1 = 3*x1+5
  c1 = -2.5*x1 + 10
  d1 = x1+11
  if y1>a1 and y1>b1 and y1<c1 and y1<d1:
    res = True
  return res
ans = checkPoint1(-1,3)
print("The point (–1, 3) is bounded with in the area enclosed by the curves of a), b), c) and d). :"+str(ans))
```

```python
def checkPoint2(x1,y1):
  res = False
  f1 = 6*x1**2 - 5*x1 + 2
  h1 = 4*x1**3 + x1**2 -3*x1 + 5
  if (y1>f1 and y1<h1) or (y1<f1 and y1>h1):
    res = True
  return res
ans2 = checkPoint2(0,3)
print('The point (0, 3) is bounded with in the area enclosed by the curves of f) and h): '+str(ans2))
```

j) f (x, y) = x + y + xy + 3




```python
i = np.linspace(-10,10,101)
j = np.linspace(-10,10,101)

X,Y = np.meshgrid(i, j)

def j(x,y):
  return x + y + x*y + 3

def k(x,y):
  return 3.5*x*y + 2*x**2+ 9*y

def l(x,y):
  return y*x**2 + 3*x*y - 7*y**2 + 2.3

def m(x,y):
  return 5*x*y +2

def n(x,y):
  return ((5*x*y) / np.exp(x**2 + y**2)) + 2

J = j(X,Y)
K = k(X,Y)
L = l(X,Y)
M = m(X,Y)
N = n(X,Y)
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, J, 100, cmap='Blues')
plt.show()
```
